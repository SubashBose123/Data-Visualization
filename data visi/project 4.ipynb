```python
import pandas as pd
data=pd.read_csv("4. covid_19_data.csv")
```

```python
data.head()
```

# show the no.of deaths,confirmed and recovered in a given region

```python
data.groupby('Region').value_counts('Confirmed','Deaths','Recovered')
```

```python
data.groupby('Region').sum()
```

```python
data.groupby("Region")['Confirmed','Deaths'].sum()
```

# remove all records were confirmed cases is less than 10

```python
data[~(data['Confirmed']<10)]
```

# in which region max no.of confirmed cases were recorded

```python
data.groupby("Region").Confirmed.sum().sort_values(ascending=False)
```

# in which region min no.of death cases were recorded.....

```python
data.groupby("Region").Deaths.sum().sort_values(ascending=True)
```
